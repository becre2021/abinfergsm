```python
def _clone_parms(ith_model):
    if hasattr(ith_model,'kernel'):
        weight_ = ith_model.kernel.weight.data.clone()
        mu_ = ith_model.kernel.mu.data.clone()
        std_ = ith_model.kernel.std.data.clone()
        var_ = ith_model.likelihood.variance.data.clone()
    return weight_,mu_,std_,var_

def _transfer_params(model_reg,weight_,mu_,std_,var_):
    model_reg.weight.data = weight_
    model_reg.mu.data = mu_
    model_reg.std.data = std_
    model_reg.likelihood.variance.data = var_
    return model_reg


def save_model_param(temp_model):
    param_dict = {}
    for ith_name,ith_param in temp_model.named_parameters():
        param_dict[ith_name] = ith_param.data.clone()    
    if hasattr(temp_model,'likelihood'):
        param_dict['var_noise'] = temp_model.likelihood.variance.data.clone()     
    return param_dict


def load_model_param(temp_model,saved_param):
    for ith_name,ith_param in temp_model.named_parameters():
        ith_param.data = saved_param[ith_name]        
    if hasattr(temp_model,'likelihood'):
        temp_model.likelihood.variance.data = saved_param['var_noise']
    return temp_model
```

```python
def _compute_kernelGramerror(x_train,y_train,x_test,y_test,
                             exact_sm,sm_norm,
                             weight_reg,
                             equal_reg,
                             naiveweight_reg,
                             naiveweight_reg2,
                             num_spt = 1,observed_period=100,current_epoch = 1):
    
    
    

    temp_error_list = []
    temp_loss_list = []
    temp_rmse_list = []
    temp_mnll_list = []
    
    assignment_list = []
    
    #for ijk in range(100):        
    for ijk in range(20):        
        
        with torch.no_grad():
#             w_loss = weight_reg._compute_nlml(x_train,y_train,intrain = False)
#             nw_loss = naiveweight_reg._compute_nlml(x_train,y_train,intrain = False)            
#             e_loss = equal_reg._compute_nlml(x_train,y_train,intrain = False)              
#             w_appkernel = weight_reg._compute_kernel_sm_approximate(x_train,intrain = False)
#             nw_appkernel = naiveweight_reg._compute_kernel_sm_approximate(x_train,intrain = False)            
#             e_appkernel = equal_reg._compute_kernel_sm_approximate(x_train,intrain = False)       

            w_loss = weight_reg._compute_nlml(x_train,y_train,intrain = True)
            e_loss = equal_reg._compute_nlml(x_train,y_train,intrain = True) 
            nw_loss = naiveweight_reg._compute_nlml(x_train,y_train,intrain = True)            
            nw_loss2 = naiveweight_reg2._compute_nlml(x_train,y_train,intrain = True)                    

#             w_loss,w_dataterm,w_logdetterm = weight_reg._compute_nlml(x_train,y_train,intrain = True)
#             nw_loss,nw_dataterm,nw_logdetterm = naiveweight_reg._compute_nlml(x_train,y_train,intrain = True)            
#             e_loss,e_dataterm,e_logdetterm = equal_reg._compute_nlml(x_train,y_train,intrain = True) 
            
            
            #assignment_list.append(naiveweight_reg.num_samplept_list_at )
            assignment_list.append(weight_reg.num_samplept_list_at )
        
        
            w_appkernel = weight_reg._compute_kernel_sm_approximate(x_train,intrain = False)
            e_appkernel = equal_reg._compute_kernel_sm_approximate(x_train,intrain = False)       
            nw_appkernel = naiveweight_reg._compute_kernel_sm_approximate(x_train,intrain = False)            
            nw_appkernel2 = naiveweight_reg2._compute_kernel_sm_approximate(x_train,intrain = False)                        

            w_pred_mu,w_pred_var = weight_reg._predict(x_test)
            e_pred_mu,e_pred_var = equal_reg._predict(x_test)                            
            nw_pred_mu,nw_pred_var = naiveweight_reg._predict(x_test)            
            nw_pred_mu2,nw_pred_var2 = naiveweight_reg2._predict(x_test)                        

            
            
            
            
            
            
            
        w_error = exact_sm - w_appkernel.cpu().data.numpy()
        e_error = exact_sm - e_appkernel.cpu().data.numpy()
        nw_error = exact_sm - nw_appkernel.cpu().data.numpy()        
        nw_error2 = exact_sm - nw_appkernel2.cpu().data.numpy()                

        w_rmse,w_mnll = _evaluate_metric(w_pred_mu,w_pred_var, y_test)
        e_rmse,e_mnll = _evaluate_metric(e_pred_mu,e_pred_var, y_test)        
        nw_rmse,nw_mnll = _evaluate_metric(nw_pred_mu,nw_pred_var, y_test)        
        nw_rmse2,nw_mnll2 = _evaluate_metric(nw_pred_mu2,nw_pred_var2, y_test)        
        

        w_error_fnorm = fnorm(w_error,'fro') 
        e_error_fnorm = fnorm(e_error,'fro')
        nw_error_fnorm = fnorm(nw_error,'fro')         
        nw_error_fnorm2 = fnorm(nw_error2,'fro')         

        
        temp_error_list.append([w_error_fnorm,e_error_fnorm,nw_error_fnorm,nw_error_fnorm2])
        temp_loss_list.append([w_loss.cpu().data.numpy(),e_loss.cpu().data.numpy(), nw_loss.cpu().data.numpy(), nw_loss2.cpu().data.numpy()])
        temp_rmse_list.append([w_rmse, e_rmse, nw_rmse, nw_rmse2])
        temp_mnll_list.append([w_mnll, e_mnll, nw_mnll, nw_mnll2])
        


    return temp_error_list,temp_loss_list,temp_rmse_list,temp_mnll_list,assignment_list






```
